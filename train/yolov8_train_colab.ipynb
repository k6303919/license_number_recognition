```python
# yolo load_model best.pt predict plate pic
model = YOLO('/content/runs/detect/train5/weights/last.pt')
img = cv2.imread('/content/content/yolov8_detection-plates-39/test/images/20200706_145200_jpg.rf.0bd7cefce45d29ce73246967ad11dac7.jpg')
#im_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # Because of OpenCV reading images as BGR
blur_gray = cv2.GaussianBlur(img,(3,3),0)
_,thresh = cv2.threshold(blur_gray, 127, 255, cv2.THRESH_BINARY_INV ) #轉為黑白
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=1)
invert = 255 - opening
results = model.predict(source=invert)
#result = results[0]



for result in results:
  boxes = result.boxes.cpu().numpy()
  for box in boxes:
    #x, y = box[0], box[1]
    #w, h = box[2], box[3]
    cords=box.xyxy[0].astype(int)
    class_id = result.names[box.cls[0].item()]
    conf = round(box.conf[0].item(), 2)
    crop = invert[int(box.xyxy[0, 1]):int(box.xyxy[0, 3]), int(box.xyxy[0, 0]):int(box.xyxy[0, 2])]
    cv2.imwrite('pred1.png', crop)
    #colors = ["green"]
    #color = str[int(c) for c in colors[class_id[0]]]
    print("Object type:", class_id)
    print("Coordinates:", cords)
    print("Probability:", conf)
    cv2.rectangle(img, cords[:2], cords[2:], (0,255,0),2)
    text = "{}: {:.2f}".format(class_id, conf)
    text2 = tool.image_to_string(
    Image.open('pred1.png'),
    builder=pyocr.builders.TextBuilder(),lang=lang)

    txt=text2.replace("|","1") # 如果是 ! 字元，更改為字元 1
    real_txt=re.findall(r'[A-Z]+|[\d]+',txt) #只取數字和大寫英文字母

    #cv2.putText(img, conf)
    cv2.putText(img, text, (cords[0], cords[1] - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 2)
    print("優化後辨識結果：",real_txt)

cv2_imshow(img)
cv2_imshow(thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
